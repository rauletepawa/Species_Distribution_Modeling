```python
import pandas as pd
import numpy as np
import pickle
import json
import matplotlib.pyplot as plt
```

```python
with open(r"..\data\Full_Scale\vectorized_gbif_assemblages_climatic_1991-2018_32.pkl", "rb") as file:
    df = pickle.load(file)
df
```

```python
# Load from JSON file
with open(r"..\data\species_universe.json", "r") as file:
    species_universe = json.load(file)
```

```python
# Convert species vectors to a NumPy array and sum along axis 0 (columns)
class_counts = np.sum(np.stack(df['species_vector'].values), axis=0)

# Convert to dictionary format {class_index: count}
class_counts_dict = {i: int(count) for i, count in enumerate(class_counts)}

# print(class_counts_dict)
sorted_dict_species_positions = dict(sorted(class_counts_dict.items(), key=lambda item: item[1], reverse = True))

species_counts = {species: class_counts_dict[position] for species, position in species_universe.items()}
sorted_dict_species = dict(sorted(species_counts.items(), key=lambda item: item[1], reverse = True))
print(sorted_dict_species)
```

```python
# # Save as JSON
# with open("../data/species_counts.json", "w") as json_file:
#     json.dump(sorted_dict_species, json_file, indent=4)

# # Save as Excel
# df = pd.DataFrame(list(sorted_dict_species.items()), columns=["Species", "Count"])
# df.to_csv("../data/species_counts.xlsx", index=False)
```

```python
# Step 1: Filter out species with count less than 10
filtered_species_positions = {k: v for k, v in sorted_dict_species_positions.items() if v >= 10}
# print(filtered_species_positions)
filtered_species = {k: v for k, v in species_counts.items() if v >= 10}
print(filtered_species)
```

```python
# Flatten the lists and remove duplicates
species_list = list(filtered_species.keys())
len(species_list),species_list
```

```python
species_universe = {}
for position, sp_id in enumerate(species_list):
    species_universe[sp_id] = position
species_universe
```

```python

# Save to JSON file
# with open("../data/species_universe_filtered_10.json", "w") as file:
#     json.dump(species_universe, file, indent=4)

with open("../data/species_universe_filtered_10.json", "r") as file:
    species_universe = json.load(file)
```

```python
#species_universe
```

```python
vector_length = len(species_universe)
# Create binary vectors for each row
def species_to_vector(species_list):
    vector = np.zeros(vector_length, dtype=int)  # Initialize zeros
    for species in species_list:
        if species in species_universe:  # Check if species is in the dictionary
            vector[species_universe[species]] = 1  # Set corresponding position to 1
    return vector

# Apply the function to create binary vectors
df.loc[:, 'species_vector'] = df['species'].apply(species_to_vector)

# Display the DataFrame
df
```
