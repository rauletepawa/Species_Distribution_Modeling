```python
def multi_label_accuracy(predictions, targets, threshold=0.1):
    """
    Computes accuracy for multi-label classification.
    Correct predictions are counted when the probability is above the threshold.
    """
    predictions = (torch.sigmoid(predictions) > threshold).float()  # Apply threshold
    correct = (predictions == targets).float()  # Compare with ground truth
    return correct.mean().item()  # Average across all labels
    
def jaccard_index(predictions, targets, threshold=0.1):
    """
    Computes the Jaccard Index (IoU) for multi-label classification.
    """
    predictions = (torch.sigmoid(predictions) > threshold).float()  # Apply threshold
    intersection = (predictions * targets).sum(dim=1)  # True positives
    union = ((predictions + targets) > 0).sum(dim=1)  # Union (all species present in either)
    return (intersection / union).mean().item()  # Mean Jaccard Index across batch
    
def top_k_accuracy(predictions, targets, k=10):
    """
    Computes top-k accuracy for multi-label classification.
    Checks if the top-k predicted species include the true species.
    """
    top_k_preds = torch.topk(torch.sigmoid(predictions), k, dim=1).indices  # Get top-k predictions
    correct = torch.zeros(predictions.shape[0], dtype=torch.float32).to(predictions.device)

    for i in range(predictions.shape[0]):  # Iterate over batch
        correct[i] = torch.any(targets[i, top_k_preds[i]] == 1).float()

    return correct.mean().item()  # Average across batch
```

```python

# Define the CNN model
class ClimaticCNN(nn.Module):
    def __init__(self, num_channels, num_classes):
        super(ClimaticCNN, self).__init__()

        # Convolutional layers with ReLU activation and MaxPooling
        self.conv1 = nn.Conv2d(in_channels=num_channels, out_channels=32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1)

        # Fully connected layers after flattening
        self.fc1 = nn.Linear(128 * 2 * 2, 512)  # Adjust according to the final feature map size
        self.fc2 = nn.Linear(512, num_classes)

        # Dropout for regularization
        self.dropout = nn.Dropout(0.3)

        # Max pooling layer
        self.pool = nn.MaxPool2d(2, 2)

    def forward(self, x):
        # Apply the layers sequentially
        x = self.pool(torch.relu(self.conv1(x)))
        # print(f'After conv1 and pool: {x.shape}')
        x = self.pool(torch.relu(self.conv2(x)))
        # print(f'After conv2 and pool: {x.shape}')
        x = self.pool(torch.relu(self.conv3(x)))
        # print(f'After conv3 and pool: {x.shape}')
        # print(x.view(x.size(0),-1).shape)
        # Flatten the output for the fully connected layers
        x = x.view(x.size(0),-1)  # Flatten, adjust the size if needed
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        
        return x


```

```python
def train_model(model, dataloader, optimizer, criterion, device):
    model.train()
    running_loss = 0.0
    total_samples = 0
    total_acc = 0
    total_jaccard = 0
    total_topk = 0

    for batch in dataloader:
        climatic_maps, species_labels = batch
        climatic_maps, species_labels = climatic_maps.to(device), species_labels.to(device)

        optimizer.zero_grad()
        outputs = model(climatic_maps)

        # Compute loss
        loss = criterion(outputs, species_labels)
        loss.backward()
        optimizer.step()

        # Update metrics
        batch_size = species_labels.size(0)
        running_loss += loss.item() * batch_size
        total_samples += batch_size
        total_acc += multi_label_accuracy(outputs, species_labels) * batch_size
        total_jaccard += jaccard_index(outputs, species_labels) * batch_size
        total_topk += top_k_accuracy(outputs, species_labels, k=10) * batch_size

    # Compute averages
    epoch_loss = running_loss / total_samples
    epoch_acc = total_acc / total_samples
    epoch_jaccard = total_jaccard / total_samples
    epoch_topk = total_topk / total_samples

    print(f'Training -> Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.4f}, Jaccard: {epoch_jaccard:.4f}, Top-10 Accuracy: {epoch_topk:.4f}')
    return epoch_loss, epoch_acc, epoch_jaccard, epoch_topk
    
def evaluate_model(model, dataloader, criterion, device):
    model.eval()  # Set the model to evaluation mode
    running_loss = 0.0
    total_samples = 0
    total_acc = 0
    total_jaccard = 0
    total_topk = 0
    # Turn on inference context manager
    with torch.inference_mode():
    # with torch.no_grad():  # No need to compute gradients
        for batch in dataloader:
            climatic_maps, species_labels = batch
            climatic_maps, species_labels = climatic_maps.to(device), species_labels.to(device)

            # Forward pass
            outputs = model(climatic_maps)

            # Compute loss
            loss = criterion(outputs, species_labels)

            # Update metrics
            batch_size = species_labels.size(0)
            running_loss += loss.item() * batch_size
            total_samples += batch_size
            total_acc += multi_label_accuracy(outputs, species_labels) * batch_size
            total_jaccard += jaccard_index(outputs, species_labels) * batch_size
            total_topk += top_k_accuracy(outputs, species_labels, k=10) * batch_size

    # Compute averages
    epoch_loss = running_loss / total_samples
    epoch_acc = total_acc / total_samples
    epoch_jaccard = total_jaccard / total_samples
    epoch_topk = total_topk / total_samples

    print(f'Validation -> Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.4f}, Jaccard: {epoch_jaccard:.4f}, Top-10 Accuracy: {epoch_topk:.4f}')
    return epoch_loss, epoch_acc, epoch_jaccard, epoch_topk


def plot_training_results(epochs, train_losses, val_losses, train_top10_accs, val_top10_accs):
    """
    Plots the training & validation loss and the top-10 accuracy over epochs.

    Args:
        epochs (int): Number of epochs.
        train_losses (list): Training loss values.
        val_losses (list): Validation loss values.
        train_top10_accs (list): Training top-10 accuracy values.
        val_top10_accs (list): Validation top-10 accuracy values.
    """
    plt.figure(figsize=(12, 5))

    # Plot training & validation loss
    plt.subplot(1, 2, 1)
    plt.plot(range(1, epochs + 1), train_losses, label="Train Loss", marker='o')
    plt.plot(range(1, epochs + 1), val_losses, label="Validation Loss", marker='o')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")
    plt.legend()
    plt.grid()

    # Plot top-10 accuracy
    plt.subplot(1, 2, 2)
    plt.plot(range(1, epochs + 1), train_top10_accs, label="Train Top-10 Accuracy", marker='o')
    plt.plot(range(1, epochs + 1), val_top10_accs, label="Validation Top-10 Accuracy", marker='o')
    plt.xlabel("Epochs")
    plt.ylabel("Top-10 Accuracy")
    plt.title("Training vs Validation Top-10 Accuracy")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()
```

```python
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction
    
    def forward(self, inputs, targets):
        # Sigmoid function to get probabilities
        probs = torch.sigmoid(inputs)
        
        # Compute the focal loss
        loss = -self.alpha * (1 - probs) ** self.gamma * targets * torch.log(probs) - \
               self.alpha * probs ** self.gamma * (1 - targets) * torch.log(1 - probs)
        
        if self.reduction == 'mean':
            return loss.mean()
        elif self.reduction == 'sum':
            return loss.sum()
        else:
            return loss

class CBFocalLoss(nn.Module):
    def __init__(self, gamma=2, alpha=1, reduction='mean'):
        super(CBFocalLoss, self).__init__()
        self.gamma = gamma  # focusing parameter
        self.alpha = alpha  # class balancing parameter
        self.reduction = reduction  # type of reduction ('mean', 'sum', or 'none')

    def forward(self, inputs, targets):
        # Sigmoid function to get probabilities from logits
        probs = torch.sigmoid(inputs)
        
        # Compute the focal loss part
        loss = -self.alpha * (1 - probs) ** self.gamma * targets * torch.log(probs) - \
               self.alpha * probs ** self.gamma * (1 - targets) * torch.log(1 - probs)
        
        # Calculate the class-balanced focal loss
        # Count the number of positive and negative samples per class
        num_pos = targets.sum(dim=0)  # sum across the batch for each species
        num_neg = targets.size(0) - num_pos  # total samples - number of positive samples
        
        # Calculate the class weights based on the number of positive/negative samples
        class_weights = num_neg / (num_pos + num_neg)  # inverse frequency balancing

        # Apply class weights to the focal loss
        loss = loss * class_weights.unsqueeze(0)  # Apply class weights to each class

        if self.reduction == 'mean':
            return loss.mean()
        elif self.reduction == 'sum':
            return loss.sum()
        else:
            return loss
```

```python
def predict_and_print(model, test_data, device, top_k=10):
    """
    Uses the trained model to make predictions on test data and prints the outputs.

    Args:
        model (torch.nn.Module): The trained model.
        test_data (DataLoader): The test dataset loader.
        device (torch.device): The device (CPU/GPU).
        top_k (int): Number of top predicted species to display.
    """
    model.eval()  # Set model to evaluation mode
    with torch.no_grad():  # No gradient tracking during inference
        for batch_idx, (inputs, labels) in enumerate(test_data):
            inputs = inputs.to(device)  # Move data to the same device as the model
            labels = labels.to(device)  # Move labels to device
            
            outputs = model(inputs)  # Get raw model outputs (logits)
            predictions = torch.sigmoid(outputs)  # Convert logits to probabilities
            print(predictions[0])
            # Get the top-k species with highest probabilities
            top_k_values, top_k_indices = torch.topk(predictions, top_k, dim=1)
            
            print(f"Batch {batch_idx + 1}:")
            for i in range(len(inputs)):
                print(f"Sample {i + 1}:")
                print(f"  Top-{top_k} Predictions: {top_k_indices[i].cpu().numpy()}")
                print(f"  Probabilities: {top_k_values[i].cpu().numpy()}")
            
            # Break after first batch if you just want a quick check
            break  
```
