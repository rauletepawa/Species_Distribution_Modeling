```python
from chelsa_cmip6.GetClim import chelsa_cmip6

chelsa_cmip6(
    activity_id='ScenarioMIP',    # The activity type (historical data)
    table_id='Amon',              # The table you are interested in (monthly data, e.g., 'Amon')
    experiment_id='ssp585',   # Use 'historical' for past data (e.g., 1900-2014)
    institution_id='MPI-M',       # The institution responsible for the model
    source_id='MPI-ESM1-2-LR',    # The source model (e.g., 'MPI-ESM1-2-LR')
    member_id='r1i1p1f1',         # The ensemble member (optional but required for some models)
    refps='1981-01-15',           # The reference period start date (e.g., Jan 1, 1981)
    refpe='2010-12-15',           # The reference period end date (e.g., Dec 31, 2010)
    fefps='2041-01-15',           # Set the future period start to match the reference period
    fefpe='2070-12-15',           # Set the future period end to match the reference period
    xmin=5.3,                     # Longitude min
    xmax=10.4,                    # Longitude max
    ymin=46.0,                    # Latitude min
    ymax=47.5,                    # Latitude max
    output='/home/raul/TFM/Data',      # Directory where you want the data saved (update with your path)
    use_esgf=False                # Set to False if not using ESGF servers
)

```
