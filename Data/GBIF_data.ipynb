```python
# comprobar que tenemos inventarios Mainland

df = pd.read_csv('GBIF/0000005-250218110819086.csv',sep='\t')
```

```python
# Group by the desired columns and aggregate the species list
grouped_species_df = df.groupby(
    ['decimalLongitude', 'decimalLatitude', 'year', 'recordedBy']
).agg({
    'species': lambda x: list(x.unique())  # Collect unique species as a list
}).reset_index()

filtered_df = grouped_species_df[grouped_species_df['species'].apply(len) >= 5]
```
